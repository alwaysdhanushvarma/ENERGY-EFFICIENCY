```python
import pandas as pd
df = pd.read_csv('/content/energy_efficiency_data.csv')
df
```

DATA INTREPRETATION


```python
y = df['Cooling_Load']
y
```

```python
x = df.drop('Cooling_Load',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_lr_train_pred = model.predict(x_train)
y_lr_test_pred = model.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score # Import r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred) # Now you can use r2_score
lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred) # Fixed a typo here, it should be lr_test_r2
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results= pd.DataFrame(['Linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
lr_results
```

```python


from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(x_train, y_train)

y_rf_train_pred = rf_model.predict(x_train)
y_rf_test_pred = rf_model.predict(x_test)

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

print(rf_results)
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

all_results = pd.concat([lr_results, rf_results])

plt.figure(figsize=(10, 6))
sns.barplot(x='Method', y='Test MSE', data=all_results)
plt.title('Comparison of Test MSE')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Method', y='Test R2', data=all_results)
plt.title('Comparison of Test R-squared')
plt.show()

print(all_results)
```
